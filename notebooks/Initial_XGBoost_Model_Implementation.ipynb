```python
import pandas as pd
import numpy as np
import collections
import joblib
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.metrics import f1_score, classification_report
from xgboost import XGBClassifier

# ----------------------------
# 1. Load dataset
# ----------------------------
df = pd.read_csv("C:\\Users\\sarav\\OneDrive\\Desktop\\Predicting_faulty_med_device\\dataset\\nlp_combinedv4.csv")

# ----------------------------
# 2. Essential features
# ----------------------------
essential_features = ["name", "name_manufacturer", "implanted", "type"]  # minimal input features
X = df[essential_features]
y = df["final_recall_level"]

# ----------------------------
# 3. Encode target labels
# ----------------------------
le = LabelEncoder()
y_enc = le.fit_transform(y)  # map 1–5 → 0–4

# ----------------------------
# 4. Preprocessing
# ----------------------------
categorical = X.select_dtypes(include=["object"]).columns.tolist()

preprocessor = ColumnTransformer(
    transformers=[
        ("cat", OneHotEncoder(handle_unknown="ignore"), categorical),
    ]
)

# ----------------------------
# 5. Train-test split
# ----------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X, y_enc, test_size=0.2, stratify=y_enc, random_state=42
)

X_train_enc = preprocessor.fit_transform(X_train)
X_test_enc = preprocessor.transform(X_test)

# ----------------------------
# 6. Compute class weights
# ----------------------------
class_counts = collections.Counter(y_train)
n_classes = len(class_counts)
n_samples = len(y_train)

class_weights = {cls: n_samples / (n_classes * count) for cls, count in class_counts.items()}
sample_weights = np.array([class_weights[label] for label in y_train])

# ----------------------------
# 7. Train XGBoost
# ----------------------------
xgb = XGBClassifier(
    tree_method="hist",
    n_estimators=300,
    learning_rate=0.1,
    max_depth=6,
    random_state=42,
    eval_metric="mlogloss"
)

xgb.fit(X_train_enc, y_train, sample_weight=sample_weights)

# Save model, preprocessor, label encoder, and class weights
joblib.dump(xgb, "xgb_model_final.pkl")
joblib.dump(preprocessor, "preprocessor_final.pkl")
joblib.dump(le, "label_encoder_final.pkl")
joblib.dump(class_weights, "class_weights_final.pkl")

# ----------------------------
# 8. Evaluate
# ----------------------------
y_pred = xgb.predict(X_test_enc)
macro_f1 = f1_score(y_test, y_pred, average="macro")
print("Macro F1 Score:", macro_f1)
print(classification_report(le.inverse_transform(y_test), le.inverse_transform(y_pred)))

# ----------------------------
# 9. Prediction function
# ----------------------------
def predict_final_recall(name, name_manufacturer, implanted, type_action):
    """
    Predict final_recall_level from minimal inputs.
    """
    # Load saved objects
    xgb_model = joblib.load("xgb_model_final.pkl")
    preproc = joblib.load("preprocessor_final.pkl")
    label_enc = joblib.load("label_encoder_final.pkl")
    
    # Prepare input DataFrame
    new_record = pd.DataFrame([{
        "name": name,
        "name_manufacturer": name_manufacturer,
        "implanted": implanted,
        "type": type_action
    }])
    
    # Preprocess
    X_new_enc = preproc.transform(new_record)
    
    # Predict
    y_pred_enc = xgb_model.predict(X_new_enc)[0]
    severity_class = label_enc.inverse_transform([y_pred_enc])[0]
    
    # Reason: check dataset for matching device & manufacturer
    match = df[(df["name"] == name) & (df["name_manufacturer"] == name_manufacturer)]
    if not match.empty:
        reason = f"Based on historical actions for this device and manufacturer. Type: {match['type'].iloc[0]}"
    else:
        reason = "No exact historical match; prediction based on model."
    
    return {
        "Predicted_Final_Recall_Level": severity_class,
        "Reason": reason,
        "Action_Type": type_action
    }

# ----------------------------
# 10. Example usage
# ----------------------------
example_prediction = predict_final_recall(
    name="Invacare Action 3 Junior Push Bar Handle",
    name_manufacturer="Invacare Poirier SAS",
    implanted="Unknown",
    type_action="Recall"
)

print(example_prediction)
```

```python
import pandas as pd
import numpy as np
import collections
import joblib
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.metrics import f1_score, classification_report
from xgboost import XGBClassifier
from fuzzywuzzy import process  

df = pd.read_csv("C:\\Users\\sarav\\OneDrive\\Desktop\\Predicting_faulty_med_device\\dataset\\nlp_combinedv4.csv")

feature_columns = ["name", "name_manufacturer", "implanted", "type",
                   "action_summary", "risk_class", "classification", "status", "determined_cause"]
X = df[feature_columns]
y = df["final_recall_level"]

le = LabelEncoder()
y_enc = le.fit_transform(y)
categorical = X.select_dtypes(include=["object"]).columns.tolist()

preprocessor = ColumnTransformer(
    transformers=[
        ("cat", OneHotEncoder(handle_unknown="ignore"), categorical),
    ]
)

X_train, X_test, y_train, y_test = train_test_split(
    X, y_enc, test_size=0.2, stratify=y_enc, random_state=42
)

X_train_enc = preprocessor.fit_transform(X_train)
X_test_enc = preprocessor.transform(X_test)

# Compute class weights
class_counts = collections.Counter(y_train)
n_classes = len(class_counts)
n_samples = len(y_train)
class_weights = {cls: n_samples / (n_classes * count) for cls, count in class_counts.items()}
sample_weights = np.array([class_weights[label] for label in y_train])

# Train XGBoost
xgb = XGBClassifier(
    tree_method="hist",
    n_estimators=300,
    learning_rate=0.1,
    max_depth=6,
    random_state=42,
    eval_metric="mlogloss"
)

xgb.fit(X_train_enc, y_train, sample_weight=sample_weights)

# Save model and preprocessing artifacts
joblib.dump(xgb, "xgb_model_hybrid.pkl")
joblib.dump(preprocessor, "preprocessor_hybrid.pkl")
joblib.dump(le, "label_encoder_hybrid.pkl")
joblib.dump(class_weights, "class_weights_hybrid.pkl")
joblib.dump(df, "full_dataset.pkl")

# 8. Evaluate model
y_pred = xgb.predict(X_test_enc)
macro_f1 = f1_score(y_test, y_pred, average="macro")
print("Macro F1 Score:", macro_f1)
print(classification_report(le.inverse_transform(y_test), le.inverse_transform(y_pred)))

# Hybrid Prediction Function
def predict_final_recall_hybrid(name, name_manufacturer, implanted, type_action,
                                 action_summary=None, risk_class=None, classification=None, status=None, determined_cause=None):
    """
    Hybrid prediction: uses dataset lookup for existing devices, XGBoost for unseen devices.
    """
    # Load saved objects
    xgb_model = joblib.load("xgb_model_hybrid.pkl")
    preproc = joblib.load("preprocessor_hybrid.pkl")
    label_enc = joblib.load("label_encoder_hybrid.pkl")
    dataset = joblib.load("full_dataset.pkl")
    
    # Fuzzy match on name
    possible_names = dataset["name"].unique()
    best_name, score_name = process.extractOne(name, possible_names)
    
    # Fuzzy match on manufacturer
    possible_manf = dataset["name_manufacturer"].unique()
    best_manf, score_manf = process.extractOne(name_manufacturer, possible_manf)
    
    # If good match
    if score_name >= 90 and score_manf >= 90:
        match = dataset[
            (dataset["name"] == best_name) &
            (dataset["name_manufacturer"] == best_manf)
        ].iloc[0]
        predicted_level = match["final_recall_level"]
        reason = f"Exact/fuzzy historical match. Type: {match['type']}, Action Summary: {match['action_summary']}"
        action_type = match['type']
        return {
            "Predicted_Final_Recall_Level": predicted_level,
            "Reason": reason,
            "Action_Type": action_type
        }
 
    # else XGBoost
    new_record = pd.DataFrame([{
        "name": name,
        "name_manufacturer": name_manufacturer,
        "implanted": implanted,
        "type": type_action,
        "action_summary": action_summary if action_summary is not None else "",
        "risk_class": risk_class if risk_class is not None else "",
        "classification": classification if classification is not None else "",
        "status": status if status is not None else "",
        "determined_cause": determined_cause if determined_cause is not None else ""
    }])
    
    X_new_enc = preproc.transform(new_record)
    y_pred_enc = xgb_model.predict(X_new_enc)[0]
    severity_class = label_enc.inverse_transform([y_pred_enc])[0]
    
    reason = "No historical match; predicted using XGBoost model."
    action_type = type_action
    
    return {
        "Predicted_Final_Recall_Level": severity_class,
        "Reason": reason,
        "Action_Type": action_type
    }

# Example usage
example_prediction = predict_final_recall_hybrid(
    name="Shear Valve Assembly",
    name_manufacturer="abbott diagnostics international ltd",
    implanted="Unknown",
    type_action="Recall"
)

print(example_prediction)
```
